# Deep Computer Vision using Convolutional Neural Networks

**Convolutional Neural Networks (CNNs)** have achieved superhuman performance on many complex visual tasks but they are not restricted to visual perception. They can be used for voice recognition and NLP, too. CNNs power things like image search, self-driving cars, and automatic video classifications. They are around since the 1980s, but only recent improvements in Deep Learning (Chapter 11) and the increased computational power allow CNNs to be so powerful.

**The inspiration for CNNs came from studies of the visual cortex**. In the late 1950s, researchers found out that many neurons in the visual cortex have a small local receptive field. Thus, they respond to small patterns in small regions of what we see. These receptive fields overlap and together make up the whole picture we see. Furthermore, there are larger receptive fields for more complex combinations of the result of lower and smaller receptive fields. This inspired the general architecture of CNNS, where higher-level neurons are based on the output of lower-level neurons. One difference to DNNs is that the layers are not fully connected to reduce the number of necessary neurons and instead detect patterns. Furthermore, in CNNs the inputs are not flat but in 2D, since the pixels are related to their neighbors.

### Convolution Layers

The most important building block for CNNs is the convolution layer. Neurons in the first convolutional layer are not connected to every single pixel in the input impage but only to pixels in their receptive fields. The neurons in the second convolutional layer are also only connected to neurons located in a small rectangle in the first layer. Thus, the network concentrates on small low-level features (e.g., an edge) in the first hidden layer and assembles them into larger high-level features (e.g., a corner) in the next hidden layer.

The receptive field glides step-wise over the input image, which is called a **stride**

The CNN detects features and represents them through the neuron's weights. These are called **filters** (e.g., vertical line, horizontal line, something complex the network found out) and we can visualize them on a **feature map**, which is a smaller image representing what the receptive field would see. All neurons in a feature map share the same parameters.

Input images are composed of multiple sublayers, one per color channel (three for RGB, one for grayscale)

### TensorFlow Implementation
In TF, each input image is typically represented as a 3D tensor of shape [height, width, channels]. 

A mini-batch is represented as a 4D tensor of shape [mini-batch size, height, width, channels].

The weights of a convolutional layer (i.e., a filter) are represented as a 4D tensor of shape [receptive field height, receptive field width, number of feature maps in previous layer, number of feature maps in current layer].

```python
import numpy as np
from sklearn.datasets import load_sample_image

# load sample images
china = load_sample_image("china.jpg") / 255
flower = load_sample_image("flower.jpg") / 255
images = np.array([china, flower])
```

```python
# create two filters
filters = np.zeros(shape=(7, 7, channels, 2), dtype=np.float32) # create 4D tensor of 0s
filters[:, 3, :, 0] = 1 # vertical line
filters[3, :, :, 1] = 1 # horizontal line
```

### Choice of Padding

If we set padding to **SAME**, the convolutional layer will use zero padding if necessary, meaning the addition of 0s around the input of the previous layer. That way, the current layer will have the same height and width of the previous layer when sliding over the input. 

Output size = Input Size / Stride

Say the input size is 13, a filter width of 6, and stride is 5, then the output size is 2.6. Thus, some pixels near the edge of the input will be ignore. Without padding, the size is reduced to 2 output pixels. However, padding allows for another stride since the input size is 13+1+1 = 15 and 15/5=3 output pixels.

When we use padding=**VALID**, the convolution layer does not use zero padding and may ignore some rows and columns at the bottom or right of the input image, depending on the stride.

```python
# create two filters
filters = np.zeros(shape=(50, 50, channels, 2), dtype=np.float32) # create 4D tensor of 0s
filters[:, 25, :, 0] = 1 # vertical line
filters[25, :, :, 1] = 1 # horizontal line
```

```python
import tensorflow as tf
outputs = tf.nn.conv2d(images, filters, strides=25, padding="SAME")
plt.imshow(outputs[0, :, :, 1], cmap="gray") # plot first image's 2nd feature map
```

```python
import tensorflow as tf
outputs = tf.nn.conv2d(images, filters, strides=25, padding="VALID")
plt.imshow(outputs[0, :, :, 1], cmap="gray") # plot first image's 2nd feature map
```

### Pooling Layers
Pooling layers allow to subsample or shrink the input layer in order to reduce the computational load, memory usage, and number of parameters. It also helps to avoid overfitting.

Just like for a convolutional layer, each neuron in a pooling layer is connected to the ouput of a limited number of neurons from the previous layer (rectangular receptive field). Again, we must define the size, the stride, and padding type. The difference is that a pooling neuron has no weights. 

Thus, the pooling neuron aggregates the input neurons using an aggregation function like max or mean.

A downside of pooling is that it's destructive by design. Even with a kernel (size) of 2x2 and a stride of 2, the resulting area will be 2^2 = 4 times smaller that in the previous layer, a reduction of information by 75%.


```python
plt.imshow(china)
```

```python
## Pooling layer: max pooling (generally found to perform better than average pooling)
outputs = tf.nn.max_pool2d(input=images, ksize=5,
                           strides=10, padding="SAME", data_format="NHWC")

plt.imshow(outputs[0, :, :, 1], cmap="gray") # plot first image's 2nd feature map
```

```python
print("Shape before pooling: " + str(china.shape))
print("Shape after pooling: " + str(outputs.shape)) # NHWC format -> 90% reduction in size
```

## CNN Architectures

In practice, CNNs stack a few convolutional layers generally followed by a ReLU layer, then a pooling layer, then another few convolutional layers, and so on. 

We shouldn't use convolutional kernels that are two large. It's better to use two 3x3 convolutional layers than one 5x5 layer. It will use fewer parameters and it will be easier to compute with a better performance. The expection might be the first convolutional layer, where a larger kernel (5x5) and a stride of 2+ reduces the spatial dimension of the image without losing too much information.

The number of filters grows throughout the CNN. The rational behind it is that there a only a few low-level features but many more ways to combine them to higher-level features. It is best practice to double the number of feature maps/filters after each pooling layer, which often divides each spatial dimension by 2. 

The image gets smaller and smaller as it progresses through the CNN, but it also gets deeper and deeper with more feature maps. 

At the end of the CNN a fully connected feedforward neural network is added as well as the final layer to output prediction results (e.g. with a softmax layer for outputting class probabilities).

### Famous Architectures

- LeNet-5 (1998)
- AlexNet (2012)
- GoogLeNet (2014)
- ResNet (2015): deep with skip connections (shortcut connections)

## Example: CNN for the MNIST dataset

```python
import tensorflow as tf
from tensorflow import keras

# fashion MNIST dataset
fashion_mnist = keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test) = fashion_mnist.load_data() #60.000 examples

class_names = ["T-Shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt",
              "Sneaker", "Bag", "Ankle boot"]
```

```python
# create validation set
X_valid, X_train = X_train_full[:5000], X_train_full[5000:]
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]

# scale input features (pixel intensity) to a 0-1 range
X_train, X_valid = X_train/255.0, X_valid/255.0
```

```python
# get input shape and add number of channels -> [55000, 28, 28, 1]
X_train = X_train[..., np.newaxis]
X_valid = X_valid[..., np.newaxis]
X_test = X_test[..., np.newaxis]
X_train.shape
```

```python
plt.imshow(X_train[1], cmap="gray")
```

```python
# define CNN
model = keras.models.Sequential([
    keras.layers.Conv2D(filters=64, 
                        kernel_size=7,
                        activation="relu", 
                        padding="same", 
                        input_shape=[28,28,1]),
    keras.layers.MaxPooling2D(2),
    keras.layers.Conv2D(128, 3, activation="relu", padding="same"),
    keras.layers.Conv2D(128, 3, activation="relu", padding="same"),
    keras.layers.MaxPooling2D(2),
    keras.layers.Conv2D(256, 3, activation="relu", padding="same"),
    keras.layers.Conv2D(256, 3, activation="relu", padding="same"),
    keras.layers.MaxPooling2D(2),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(10, activation="softmax")
])
```

```python
# print model summary
model.summary()
```

```python
# plot model
# keras.utils.plot_model(model, show_shapes=True)
```

```python
# specify loss function and optimizer
model.compile(loss="sparse_categorical_crossentropy",
             optimizer="nadam",
             metrics=["accuracy"])
```

```python
%%time
# fit the model
history = model.fit(X_train, y_train, epochs=10, 
                    validation_data=(X_valid, y_valid))
```

```python
score = model.evaluate(X_test, y_test)
score
```

```python
y_test_pred = model.predict(X_test)
y_test_pred = np.argmax(y_test_pred, axis=1) # from matrix to predicted category
y_test_pred
```

```python
y_test
```

```python
# plot confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_test_pred)

import seaborn as sns
import matplotlib.pyplot as plt
plt.subplots(figsize=(10,8)) 
sns.heatmap(confusion_matrix, xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix for MNIST Fashion", size=16)
```

## Implementing ResNet-34 CNN using Keras (CIFAR-10)

A Residual Network (ResNet) is a deep convolutional neural network (for example with 34 convolutional layers). These networks are able to be trained using **skip or shortcut connections** which feed the output directly to a layer further down in the network. The ResNet can be seen as a network consisting of **residual units (RUs)** where each RU is a small network with a skip connection.

Each RU consists of two convolutional layers with batch normalization and ReLU activation, using 3x2 kernels and a stride of 1 with padding, to preserve spatial dimensions. No pooling layer.

First we define the residual units and skip connections. Then we chain then together using a Sequential model

```python
import tensorflow as tf
from tensorflow import keras

# CIFAR-10 dataset
cifar10 = keras.datasets.cifar10
(X_train_full, y_train_full), (X_test, y_test) = cifar10.load_data() #50.000 examples

class_names = ["Airplane", "Automobile", "Bird", "Cat", "Deer", "Dog", "Frog", "Horse", "Ship", "Frog"]
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train_full[0])
```

```python
X_train_full.shape
```

```python
# resize image for ResNet 224x224x3 (upsampling)
X_train_full_resized = tf.image.resize(X_train_full, [224, 224], antialias=True, method="bicubic")
X_train_full_resized.shape
```

```python
X_train_full_resized = tf.cast(X_train_full_resized, "uint8")
```

```python
plt.figure(figsize=(4,4))
plt.imshow(X_train_full_resized[0])
```

```python
# create validation set
X_valid, X_train = X_train_full_resized[:5000], X_train_full_resized[5000:]
y_valid, y_train = X_train_full_resized[:5000], X_train_full_resized[5000:]

# keep input features (pixel intensity) on a 0-255 range
# X_train[0,:]

# check shape
X_train.shape
```

```python
# This model won the ILSVRC 2015 challenge and is very powerful, yet "simply" coded in less than 40 lines

import tensorflow as tf
from tensorflow import keras


# Define RUs
class ResidualUnit(keras.layers.Layer):
    def __init__(self, filters, strides=1, activation="relu", **kwargs):
        super().__init__(**kwargs)
        self.activation = keras.activations.get(activation)
        self.main_layers = [
            keras.layers.Conv2D(filters, 3, strides=strides,
                               padding="same", use_bias=False),
            keras.layers.BatchNormalization(),
            self.activation,
            keras.layers.Conv2D(filters, 3, strides=1, padding="same", use_bias=False),
            keras.layers.BatchNormalization()
        ]
        self.skip_layers = []
        if strides > 1:
            self.skip_layers = [
                keras.layers.Conv2D(filters, 1, strides=strides,
                                   padding="same", use_bias=False),
                keras.layers.BatchNormalization()
            ]
    
    def call(self, inputs):
        Z = inputs
        for layer in self.main_layers:
            Z = layer(Z)
        skip_Z = inputs
        for layer in self.skip_layers:
            skip_Z = layer(skip_Z)
        return self.activation(Z + skip_Z)
```

```python
# Define Sequential model and build ResNet-34
model = keras.models.Sequential()
model.add(keras.layers.Conv2D(64, 7, strides=2, input_shape=[224, 224, 3], padding="same", use_bias=False))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Activation("relu"))
model.add(keras.layers.MaxPool2D(pool_size=3, strides=2, padding="same"))
prev_filters = 64

# add 16 RUs
for filters in [64] * 3 + [128] * 4 + [256] * 6 + [512] * 3:
    strides = 1 if filters == prev_filters else 2
    model.add(ResidualUnit(filters, strides=strides))
    pre_filters = filters

# end with a dense model and output layer
model.add(keras.layers.GlobalAvgPool2D())
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(10, activation="softmax"))
```

```python
# print model summary
model.summary()
```

```python
# specify loss function and optimizer
model.compile(loss="sparse_categorical_crossentropy",
             optimizer="nadam",
             metrics=["accuracy"])
```

```python
%%time
# fit the model
history = model.fit(X_train, y_train, epochs=10, 
                    validation_data=(X_valid, y_valid))
```

```python
score = model.evaluate(X_test, y_test)
score
```

```python
# predict 
y_test_pred = model.predict(X_test)
y_test_pred = np.argmax(y_test_pred, axis=1) # from matrix to predicted category

# plot confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_test_pred)

import seaborn as sns
import matplotlib.pyplot as plt
plt.subplots(figsize=(10,8)) 
sns.heatmap(confusion_matrix, xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix for MNIST Fashion using modified ResNet-34", size=16)
```

## Pre-Trained Model from Keras: ResNet-50 and CIFAR10

The keras.applications package allows us to load pre-defined and pre-trained networks with a single line of code. For example, ResNet-50 trained on ImageNet.

These models assume thes at t pre-processed in a certain way. The preprocess_input() function can be used for that.

In this example we will load a ResNet-50 model trained on ImageNet to predict on CIFAR10 data with and without transfer learning.
